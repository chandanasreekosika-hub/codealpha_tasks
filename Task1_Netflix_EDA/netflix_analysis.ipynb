```python
import pandas as pd
```

```python
df = pd.read_csv('../dataset/netflix_titles.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
df['type'].value_counts().plot(kind='bar')
plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.savefig('../images/movies_vs_tvshows_bar.png')
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='bar')
plt.title('Top 10 Countries with Most Netflix Content')
plt.xlabel('Country')
plt.ylabel('Number of Shows/Movies')
plt.xticks(rotation=45)
plt.savefig('../images/ratings_distribution.png')
plt.show()
```

```python
year_data = df['release_year'].value_counts().sort_index()
year_data.plot(kind='line', figsize=(10,5))
plt.title('Netflix Content Released Over Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Shows/Movies')
plt.savefig('../images/release_trend.png')
plt.show()
```

```python
df['rating'].value_counts().head(10).plot(kind='bar')
plt.title('Top Netflix Content Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.savefig('../images/ratings_distribution.png')
plt.show()
```

```python
df['type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)
plt.title('Movies vs TV Shows Distribution')
plt.ylabel('')
plt.savefig('../images/movies_tvshows_pie.png')
plt.show()
```

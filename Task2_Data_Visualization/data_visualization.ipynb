```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../dataset/Sample - Superstore.csv', encoding='latin1')

df.head()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

plt.figure(figsize=(8,5))
category_sales.plot(kind='bar')

plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')

plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

plt.figure(figsize=(8,5))
category_sales.plot(kind='bar')

plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')

plt.savefig('../images/category_sales.png')
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum()

plt.figure(figsize=(8,5))
region_sales.plot(kind='pie', autopct='%1.1f%%')

plt.title('Sales Distribution by Region')
plt.ylabel('')

plt.savefig('../images/region_sales_pie.png')
plt.show()
```

```python
profit_by_category = df.groupby('Category')['Profit'].sum()

plt.figure(figsize=(8,5))
profit_by_category.plot(kind='bar', color='green')

plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')

plt.savefig('../images/profit_by_category.png')
plt.show()
```

```python
top_subcategories = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
top_subcategories.plot(kind='bar')

plt.title('Top 10 Sub-Categories by Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Sales')

plt.savefig('../images/top10_subcategories.png')
plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(df['Sales'], df['Profit'])

plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')

plt.savefig('../images/sales_vs_profit.png')
plt.show()
```

```python
corr = df[['Sales', 'Quantity', 'Discount', 'Profit']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.savefig('../images/correlation_heatmap.png')
plt.show()
```

# Superstore Data Visualization Project
## Objective
Analyze Superstore sales data using various visualizations to identify trends, sales performance, regional distribution, and profit patterns.
## Tools Used
- Python
- Pandas
- Matplotlib
- Seaborn
## Visualizations Created
1. Sales by Category (Bar Chart)
2. Sales by Region (Pie Chart)
3. Profit by Category
4. Top 10 Subcategories
5. Sales vs Profit (Scatter Plot)
6. Correlation Heatmap

## Key Insights
- Technology category generated the highest sales.
- West region contributed the largest share of sales.
- Some high-sales orders resulted in losses.
- Discounts negatively impacted profit in several cases.
- Certain subcategories significantly outperformed others.
